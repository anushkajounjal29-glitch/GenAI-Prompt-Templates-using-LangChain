```python

# INSTALL & IMPORT (UPDATED)

!pip install -U langchain langchain-core

from langchain_core.prompts import PromptTemplate, ChatPromptTemplate



# TASK 1: Replace Hardcoded Prompt

template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print("\n--- Task 1 Output ---")
print(template_1.format(topic="Artificial Intelligence"))



# TASK 2: Multi-Input Prompt System

template_2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print("\n--- Task 2 Output ---")
print(template_2.format(topic="AI", audience="beginners", tone="friendly"))
print(template_2.format(topic="Python", audience="kids", tone="fun"))
print(template_2.format(topic="Deep Learning", audience="engineers", tone="technical"))



# TASK 3: Prompt Variations Engine

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print("\n--- Task 3 Output ---")
print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))



# TASK 4: ChatPromptTemplate System

def chat_prompt(role, topic):

    system_messages = {
        "teacher": "You are a helpful teacher.",
        "interviewer": "You are a job interviewer.",
        "motivator": "You are a motivational speaker."
    }

    system_template = system_messages.get(role, "You are a helpful assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_template),
        ("user", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)

print("\n--- Task 4 Output ---")
messages = chat_prompt("teacher", "Neural Networks")
for msg in messages:
    print(msg)



# TASK 5: Input Validation Layer

def validate_inputs(audience, tone):

    valid_audience = ["beginners", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginners'")
        audience = "beginners"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'formal'")
        tone = "formal"

    return audience, tone

print("\n--- Task 5 Output ---")
print(validate_inputs("beginners", "fun"))        # valid
print(validate_inputs("kids", "cool"))            # invalid



# TASK 6: Prompt Generator App

def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    selected_template = templates.get(style)

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template
    )

    return prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

print("\n--- Task 6 Output ---")
print(generate_prompt("Neural Networks", "beginners", "fun", "storytelling"))



# TASK 7: Template Reusability Test

reusable_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

topics = ["AI", "Blockchain", "Cloud Computing", "Cybersecurity", "Data Science"]

print("\n--- Task 7 Output ---")
for t in topics:
    print(reusable_template.format(topic=t))



# FINAL PIPELINE FLOW

def prompt_pipeline(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    result = generate_prompt(topic, audience, tone, style)

    return result

print("\n--- Final Pipeline Output ---")
print(prompt_pipeline("AI", "beginners", "fun", "teaching"))
```
